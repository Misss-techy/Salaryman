```python
#import necessary libraries
import pandas as pd
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
import selenium
from selenium import webdriver
```

```python
driver_path='C:/Users/sumai/Documents/chromedriver'
driver = webdriver.Chrome(executable_path = driver_path)
```

```python
#Create dataframe

df1 = pd.DataFrame(columns=["Title","Company","Location","Salary","Sponsored","Summary"])
```

```python
for i in range(0,500,10):
    driver.get("https://in.indeed.com/jobs?q=Computer+engineer&l=Chennai%2C+Tamil+Nadu&start="+str(i))
    driver.implicitly_wait(4)

    all_jobs = driver.find_elements_by_class_name('result')
    for job in all_jobs:
        result_html = job.get_attribute('innerHTML')
        soup = BeautifulSoup(result_html,'html.parser')
        try:
            Title = soup.find("a",class_="jobtitle").text.replace('\n','')
        except:
            Tilte =''
        try:
            Location = soup.find(class_="location").text
        except:
            Location =''
        try:
            Company = soup.find(class_="company").text.replace("\n","").strip()
        except:
            Company =''  
        try:
            Salary = soup.find(class_="salary").text.replace('\n','').strip()
        except:
            Salary=''
        try:
            Sponsored= soup.find(class_="remote").text
            Sponsored = "Remote"
        except:
            Sponsored='In-Office'
        
        sum_div = job.find_elements_by_class_name("summary")[0]
        try:
            sum_div.click()
        except:
            close_button = driver.find_elements_by_class_name("popover-x-button-close icl-CloseButton")
            close_button.close()
            sum_div.click()
        try:   
            job_desc= driver.find_element_by_id('vjs-desc').text
        except: 
            job_desc=''
            
        df1 =df1.append({'Title':Title,'Company':Company,'Location':Location,'Salary':Salary,'Sponsored':Sponsored,'Summary':job_desc},ignore_index=True)
    
```
